```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
DATA = pd.DataFrame(pd.read_csv('heart.csv'))
```

```python
DATA.keys()
```

```python
DATA.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
DATA['sex'] = le.fit_transform(DATA['Sex'])
DATA['chestpaintype'] = le.fit_transform(DATA['ChestPainType'])
DATA['Exercise_angina'] = le.fit_transform(DATA['ExerciseAngina'])
DATA['restingECG'] = le.fit_transform(DATA['RestingECG'])
DATA['ST_slope'] = le.fit_transform(DATA['ST_Slope'])
```

```python
x = DATA[['Age', 'RestingBP', 'Cholesterol', 'FastingBS', 'MaxHR', 'Oldpeak', 'sex', 'chestpaintype', 'restingECG', 'ST_slope']]
```

```python
x.head()
```

```python
y = DATA['HeartDisease']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 30)
x_train
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(x_train, y_train)
model.score(x_test, y_test)
```

```python
from sklearn.decomposition import PCA
pc = PCA(0.80)
x_pc = pc.fit_transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_pc, y, test_size = 0.2, random_state = 30)
```

```python
model.fit(x_train, y_train)
model.score(x_test, y_test)
```

```python
pc_2 = PCA(n_components=5)
x_pc2 = pc.fit_transform(x)
x_train, x_test, y_train, y_test = train_test_split(x_pc2, y, test_size = 0.2, random_state = 30)
model.fit(x_train, y_train)
model.score(x_test, y_test)
```
